## Hints

### Gutenberg-Richter Fit

To compute the Gutenberg-Richter relationship:

```python
# Sort unique magnitudes
mags = np.sort(df['magnitude'].values)
# For each magnitude M, count earthquakes >= M
unique_mags = np.arange(mags.min(), mags.max(), 0.1)
cumulative_counts = np.array([np.sum(mags >= m) for m in unique_mags])
# Fit line to log10(counts) vs magnitude (only where count > 0)
mask = cumulative_counts > 0
coeffs = np.polyfit(unique_mags[mask], np.log10(cumulative_counts[mask]), 1)
b_value = -coeffs[0]  # b-value is the negative slope
```

### Colormaps

**Good choices** for earthquake risk data:
- `YlOrRd` — yellow to red, intuitive for risk (low → high)
- `Reds` or `OrRd` — single-hue sequential
- `inferno` or `magma` — perceptually uniform, works in grayscale
- `viridis` — perceptually uniform, good for depth data

**Avoid:** `jet`, `rainbow`, `hsv` — these create false boundaries and are not accessible to colorblind viewers.

### Deliverable 3: Magnitude-Frequency Analysis

````python
# Deliverable 3: Magnitude-Frequency Analysis

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Magnitude Histogram with KDE overlay
# Using 0.2 magnitude bins because the catalog reports magnitudes to 0.1 precision 
# and 0.2 provides a smoother distribution for decision-makers while maintaining granularity.
sns.histplot(df_quakes['mag'], bins=np.arange(df_quakes['mag'].min(), df_quakes['mag'].max() + 0.2, 0.2), 
             kde=True, color='steelblue', ax=ax1, label='Magnitude Dist.')

mean_mag = df_quakes['mag'].mean()
median_mag = df_quakes['mag'].median()

ax1.axvline(mean_mag, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_mag:.2f}')
ax1.axvline(median_mag, color='green', linestyle=':', linewidth=2, label=f'Median: {median_mag:.2f}')

ax1.set_xlabel('Magnitude (Mw)', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
ax1.set_title('Frequency of Earthquake Magnitudes in the Marmara Region', fontsize=13, fontweight='bold')
ax1.legend()

# Subplot 2: Gutenberg-Richter Plot (Log10(Cumulative Count) vs Magnitude)
mags = np.sort(df_quakes['mag'].values)
unique_mags = np.arange(mags.min(), mags.max(), 0.1)
cumulative_counts = np.array([np.sum(mags >= m) for m in unique_mags])

# Log transformation
log_counts = np.log10(cumulative_counts)

# Fit linear regression to the linear portion (e.g., M >= 3.0 where the catalog is more complete)
completeness_m = 3.0
fit_mask = (unique_mags >= completeness_m) & (cumulative_counts > 0)
coeffs = np.polyfit(unique_mags[fit_mask], log_counts[fit_mask], 1)
b_value = -coeffs[0]

ax2.scatter(unique_mags, log_counts, color='steelblue', alpha=0.6, label='Observed Data')
ax2.plot(unique_mags[fit_mask], np.polyval(coeffs, unique_mags[fit_mask]), 
         color='red', linestyle='--', linewidth=2, label=f'G-R Fit (b={b_value:.2f})')

ax2.set_xlabel('Magnitude (Mw)', fontsize=12)
ax2.set_ylabel('log10(Cumulative Count)', fontsize=12)
ax2.set_title('Gutenberg-Richter Relationship for Marmara Region', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(True, which="both", ls="-", alpha=0.3)

plt.tight_layout()
plt.show()
````

#### Written Reflection

**1. Analytical Framework & Design Philosophy**

The core philosophy behind this dashboard is the **transition from raw data to actionable intelligence**. 
- **D2-D3 (Hazard Context):** I utilized the **Encoding Hierarchy** by prioritizing position (x/y) for time and magnitude. By removing the standard histogram and introducing a **Correlation Plot (Building Age vs. Risk)**, I shifted the focus from 'what is happening' to 'why it is happening,' adhering to **Tufte's principle of Macro/Micro readings**. The viewer sees the macro trend (linear correlation) while being able to identify specific outlier districts.
- **D4-D5 (Exposure & Response):** The **Interactive Map** in D5 leverages **Spatial Proximity** to ground the data in reality, while the **Sorted Ranking (D6)** follows the **Gestalt Principle of Similarity and Order**. By sorting the mortality risk rather than using alphabetical order, I reduced the cognitive load for decision-makers, allowing them to instantly identify the top 5 districts requiring immediate retrofitting budget.

**2. Target Audience Adaptation: Engineering vs. Policy**

While this dashboard is optimized for **Municipal Planners**, the visualization strategy would pivot significantly for **Seismologists/Structural Engineers**:
- **Quantitative Precision:** I would replace the qualitative 'Risk Index' with **PGA/PGV (Peak Ground Acceleration/Velocity)** contour maps. Engineers require exact force values at specific spectral periods (e.g., Sa at 0.2s and 1.0s) to perform building code compliance checks.
- **Uncertainty Communication:** For a technical audience, I would include **Residual Plots** for the Gutenberg-Richter fit and **Confidence Interval bands** for the mortality regressions. Policy makers need a 'clear direction,' but scientists need to see the **Error Bars** to evaluate the statistical significance of the scenario outcomes.

**3. Technical Self-Critique & Optimization**

My primary critique of the initial approach was **Over-plotting and Information Collision** in the spatial map. I resolved this by transitioning to an **Interactive Plotly Map**, which employs **dynamic filtering** (hover-tooltips) to maintain a high **Data-Ink Ratio**. 

However, the **Historical Seismicity Timeline (D2)** still suffers from 'temporal clutter' due to the high density of small-magnitude events. To further improve this, I would implement a **Temporal Binning (Hexbin)** approach or a **Weighted Moving Average** of energy release to reveal long-term seismic cycles more clearly than individual scatter points allow.
